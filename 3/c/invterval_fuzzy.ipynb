```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def falling(x, a, b):
    if x <= a:
        return 1
    elif a < x <= b:
        return (b - x) / (b - a)
    else:
        return 0
    
def falling_min(x, a, b, min):
    if x <= a:
        return min
    elif a < x <= b:
        return (x-b)*min/(a-b)
    else:
        return 0
    
def triangular(x, a, b, c):
    if x <= a:
     return 0
    elif a < x <= b:
        return (x - a) / (b - a)
    elif b < x <= c:
        return (c - x) / (c - b)
    else:
      return 0

def triangular_min(x, a, b, c, min):
    if x <= a:
     return 0
    elif a < x <= b:
        return (x-a)*min/(b-a)
    elif b < x <= c:
        return (x-c)*min/(b-c)
    else:
      return 0
    
def trapezoidal(x, a, b, c, d):
    if x <= a:
     return 0
    elif a < x <= b:
        return (x - a) / (b - a)
    elif b < x <= c:
        return 1
    elif c < x <= d:
        return (d - x) / (d - c)
    else:
        return 0
    
def trapezoidal_min(x, a, b, c, d, min):
    if x <= a:
     return 0
    elif a < x <= b:
        return (x-a)*min/(b-a)
    elif b < x <= c:
        return min
    elif c < x <= d:
        return (x-d)*min/(c-d)
    else:
        return 0
def rising(x, a, b):
    if x <= a:
        return 0
    elif a < x <= b:
        return (x - a) / (b - a)
    else:
        return 1

def rising_min(x, a, b, min):
    if x <= a:
        return 0
    elif a < x <= b:
        return (x-a)*min/(b-a)
    else:
        return min
```

```python
players = pd.read_csv("all_seasons.csv")
```

```python
players = players[["player_name", "age", "player_height", "player_weight"]]
players['young_age_min'] = players['age'].apply(lambda age: falling_min(age, 20, 40, 0.8))
players['young_age_max'] = players['age'].apply(lambda age: falling(age, 20, 40))

players['medium_age_min'] = players['age'].apply(lambda age: trapezoidal_min(age, 30,40,45,55, 0.8))
players['medium_age_max'] = players['age'].apply(lambda age: trapezoidal(age, 30,40,45,55))

players['old_age_min'] = players['age'].apply(lambda age: rising_min(age, 50,65, 0.8))
players['old_age_max'] = players['age'].apply(lambda age: rising(age, 50,65))
```
